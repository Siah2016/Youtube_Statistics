### Imports

```python
import pandas as pd 
```

### Creating DataFrame from CSV file

```python
df = pd.read_csv("Global YouTube Statistics.csv", encoding=('ISO=8859-1'), low_memory=False, index_col="rank")

df.head()
```

#### Information on nulls

```python
df.info()
```

```python
non_null_df = df.dropna()

non_null_df.info()
```

### Delete Columns

```python
removal_col_df = non_null_df.drop(["Title", "uploads", "Abbreviation", "created_month", "created_date", "Unemployment rate", "subscribers_for_last_30_days", "Urban_population", "Latitude", "Longitude"], axis=1)

removal_col_df.info()
```

### Checking for duplicates

```python
removal_col_df.duplicated()
```

#### Checking to columns for what is needed

```python
removal_col_df.head()
```

### Cont. Removing Columns

```python
removal_2_df = removal_col_df.drop(["video views", "video_views_rank", "country_rank", "channel_type_rank", "Population", "video_views_for_the_last_30_days", "lowest_monthly_earnings", "lowest_yearly_earnings"], axis=1)

removal_2_df.head()
```

### Identifying Outlier Values

```python
sorted_df = removal_2_df.sort_values("Gross tertiary education enrollment (%)")

sorted_df.tail()
```

### Removing Outlier Values

```python
cleaned_df = sorted_df.drop([479,786])

cleaned_df.tail()
```

### Renaming Colums

```python
renamed_df = cleaned_df.rename(
    columns={
        "highest_monthly_earnings": "Monthly_Earnings",
        "highest_yearly_earnings": "Yearly_Earnings",
        "Gross tertiary education enrollment (%)": "College_Enrollment_Percentage"})

renamed_df.head()
```

#### Reformatting Columns

```python
uppercase_df = renamed_df.rename(
    columns={
        "subscribers": "Subscribers",
        "category": "Category",
        "channel_type": "Channel_Type",
        "created_year": "Created_Year"})

uppercase_df.head()
```

### Exporting to CSV file

```python
uppercase_df.to_csv("clean_youtube_statistics.csv", encoding="utf-8")
```
